### Data Preprocessing

#### Converting target variable (Attrition) to binary 

```python
a['Attrition'] = a['Attrition'].map({'Yes':1,'No':0})
```

```python
a.head()
```

#### Finding unique values for Education Fields

```python
print(a.Education.unique())
print(a.EducationField.unique())
```

```python
a['Education'] = a['Education'].map({1: 'High School',2:'Undergrad',3:'Graduate',4:'Post Graduate',5:'Doctorate'})
```

#### Finding unique values for Rating Factor columns

```python
print(a.JobInvolvement.unique())
print(a.JobSatisfaction.unique())
print(a.PerformanceRating.unique())
```

```python
cols = ["JobInvolvement", "JobSatisfaction", "PerformanceRating"]

for col in cols:
    a[col].replace({1 : "Low",2 : "Medium",3 : "High",4 : "Very High"}, inplace = True)
a.head()
```

### Attrition by Specific Job Role

```python
plt.figure(figsize=(20,20))
pd.crosstab(a.JobRole,a.Attrition).plot(kind='barh')
```

```python
c = pd.crosstab(a.JobRole,a.Attrition)
c['Attrition_Ratio']=c[1]/c[0]
c.sort_values(by=['Attrition_Ratio'],ascending=False)
```

- Sales Respresentative has the highest attrition rate while Research Director has low Attrition Rate

```python
plt.figure(figsize=(20,10))
sb.boxplot(a.JobRole,a.MonthlyIncome)
```
